#1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.


- Anomaly Detection is a data analysis technique used to identify data points, patterns, or observations that significantly deviate from the normal or expected behavior of a dataset. These unusual observations are called anomalies or outliers and may indicate errors, rare events, fraud, faults, or novel patterns.

- Types of Anomalies:
1. Point Anomaly

A point anomaly is a single data instance that is significantly different from the rest of the data.

- Example:

- In a dataset of daily bank transactions, if most transactions are between ₹500 and ₹10,000, a transaction of ₹5,00,000 can be considered a point anomaly.

- In temperature data, a sudden reading of 60°C on a normal day can be a point anomaly.

2. Contextual Anomaly

A contextual anomaly is a data point that is anomalous only in a specific context such as time, location, or season.

- Example:

- A temperature of 30°C is normal in summer but anomalous in winter.

- High electricity consumption during the daytime may be normal, but the same consumption at midnight can be anomalous.

- 3. Collective Anomaly

A collective anomaly occurs when a group of related data points together deviates from normal behavior, even if individual points seem normal.

- Example:

- A sequence of continuous failed login attempts may indicate a cyber attack.

- A sudden pattern of gradual increase in network traffic over time may indicate a DDoS attack.

#3: What are the key components of a Time Series? Explain each with one example.
- Answer:
A time series is a sequence of data points recorded at regular time intervals. It consists of the following four key components:

1. Trend (T)-
The trend represents the long-term upward or downward movement in the data over time.

- Example:
A steady increase in monthly mobile phone users over several years shows an upward trend.

- 2. Seasonality (S)-
Seasonality refers to repeating patterns or fluctuations that occur at fixed intervals (daily, monthly, yearly).

- Example:
Electricity consumption increases every summer due to air conditioner usage.

- 3. Cyclical Component (C)-
The cyclical component represents long-term fluctuations caused by economic or business cycles. These patterns do not have a fixed period.

- Example:
An increase and decrease in company profits due to economic boom and recession.

- 4. Irregular / Random Component (R)-
  This component represents unpredictable, random variations that cannot be explained by trend, seasonality, or cycles.

- Example:
A sudden drop in sales due to a natural disaster or pandemic.

# 4: Define Stationarity in time series. How can you test and transform a non-stationary series into a stationary one? (Concise Version)

- A stationary time series has statistical properties (mean, variance, autocorrelation) that remain constant over time. This is crucial for many time series models.

## How to Test for Stationarity?
1.  Visual Inspection: Plot data to check for trends, seasonality, or changing variance (e.g., rolling mean/standard deviation plots, ACF/PACF plots).
2.  Statistical Tests:
    * Augmented Dickey-Fuller (ADF) Test**: Tests if a series is non-stationary (null hypothesis). Rejecting the null suggests stationarity.
    *   Kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test: Tests if a series is stationary (null hypothesis). Rejecting the null suggests non-stationarity.

## How to Transform to Stationarity?
1.  Differencing: Subtracting previous observations (Y_t = X_t - X_{t-1}) to remove trends and seasonality. Seasonal differencing handles seasonal patterns.
2.  Log Transformation: Stabilizes variance, especially when variance increases with the mean (log(X_t)).
3.  Other Power Transformations: Similar to log, can help stabilize variance and make distributions symmetric.
4.  Seasonal Decomposition: Separates the series into trend, seasonal, and residual components; the residual is often more stationary.

# 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.


---

- 1. AR (Autoregressive) Model

*   Structure: An AR(p) model predicts the current value based on a linear combination of its past 'p' values. It captures the dependency between an observation and a number of lagged observations.
    *   Equation: Y_t = c + φ_1*Y_{t-1} + φ_2*Y_{t-2} + ... + φ_p*Y_{t-p} + ε_t
*   Application: Best suited for stationary time series where the current value is primarily influenced by its own previous values (e.g., stock prices where today's price is highly dependent on yesterday's).

---

- 2. MA (Moving Average) Model

*   Structure: An MA(q) model predicts the current value based on a linear combination of past 'q' forecast errors (the difference between the actual value and the predicted value at past time points).
    *   Equation: Y_t = c + ε_t + θ_1*ε_{t-1} + θ_2*ε_{t-2} + ... + θ_q*ε_{t-q}
*   Application: Useful for stationary time series where the current value is influenced by random shocks or unforeseen events in the past (e.g., sudden changes in demand that persist for a short period).

---

- 3. ARIMA (Autoregressive Integrated Moving Average) Model

*   Structure: ARIMA(p, d, q) combines the AR and MA components with an 'I' (Integrated) component.
    *   `AR(p)`: Autoregressive part, 'p' is the order of the AR component.
    *   `I(d)`: Integrated part, 'd' is the degree of differencing required to make the series stationary (removes trends).
    *   `MA(q)`: Moving Average part, 'q' is the order of the MA component.
*   Application: Ideal for non-stationary time series that exhibit a trend. Differencing transforms the series into a stationary one before applying AR and MA components. Widely used for general time series forecasting when seasonality is not dominant.

---

- 4. SARIMA (Seasonal Autoregressive Integrated Moving Average) Model

*   Structure: SARIMA(p, d, q)(P, D, Q)m extends ARIMA by adding seasonal components.
    *   (p, d, q): Non-seasonal AR, I, MA orders.
    *   (P, D, Q)m: Seasonal AR, I, MA orders, where 'm' is the number of observations per season (e.g., 12 for monthly data, 4 for quarterly data).
    *   P: Seasonal autoregressive order.
    *   D: Seasonal differencing order.
    *   Q: Seasonal moving average order.
*   Application: Crucial for time series data that exhibit clear seasonal patterns in addition to non-seasonal trends and cycles (e.g., monthly sales data with yearly seasonality).

---

-  5. SARIMAX (Seasonal Autoregressive Integrated Moving Average with Exogenous Regressors) Model

*   Structure: SARIMAX(p, d, q)(P, D, Q)m [exog] further extends SARIMA by incorporating exogenous variables (external factors).
    *   It has all the components of SARIMA.
    *   exog: Represents one or more independent variables that influence the time series but are not part of the time series itself.
*   Application: Used when external factors significantly influence the time series, in addition to its own historical patterns and seasonality. For example, predicting electricity consumption using SARIMA components plus external variables like temperature or holiday indicators.

#6:Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
data = pd.read_csv(
    "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv",
    parse_dates=["Month"],
    index_col="Month"
)

plt.figure()
plt.plot(data, label="Air Passengers")
plt.title("Original AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend()
plt.show()

decomposition = seasonal_decompose(data, model="multiplicative")

decomposition.plot()
plt.show()
```

#7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

np.random.seed(42)
X = np.random.randn(300, 2)
outliers = np.random.uniform(low=-6, high=6, size=(20, 2))
X = np.vstack([X, outliers])
iso = IsolationForest(contamination=0.06)
labels = iso.fit_predict(X)

normal = X[labels == 1]
anomaly = X[labels == -1]

plt.figure()
plt.scatter(normal[:, 0], normal[:, 1], label="Normal Data")
plt.scatter(anomaly[:, 0], anomaly[:, 1], label="Anomalies")
plt.title("Isolation Forest Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

#8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
data = pd.read_csv(
    "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv",
    parse_dates=["Month"],
    index_col="Month"
)

model = SARIMAX(
    data,
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12)
)
results = model.fit()
forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
plt.figure()
plt.plot(data, label="Original Data")
plt.plot(forecast_mean, label="12-Month Forecast")
plt.title("SARIMA Forecast of AirPassengers")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend()
plt.show()
```

#9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

np.random.seed(42)
X = np.random.randn(300, 2)
outliers = np.random.uniform(low=-6, high=6, size=(20, 2))
X = np.vstack([X, outliers])

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.06)
labels = lof.fit_predict(X)

normal = X[labels == 1]
anomaly = X[labels == -1]

plt.figure()
plt.scatter(normal[:, 0], normal[:, 1], label="Normal Data")
plt.scatter(anomaly[:, 0], anomaly[:, 1], label="Anomalies")
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
